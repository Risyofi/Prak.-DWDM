# Latihan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris.shape
```

```python
iris['Species'].unique()
```

```python
iris.info()
```

```python
iris.drop(columns='Id', inplace=True)
iris.isnull().sum()
```

```python
import missingno as msno
msno.bar(iris, figsize=(8,6), color='skyblue')
plt.show()
```

```python
g = sns.relplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=iris)
g.fig.set_size_inches(10, 5)
plt.show()
```

```python
iris.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['Species'] = le.fit_transform(iris['Species'])
iris.head()
```

```python
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
x = iris.drop('Species', axis=1)
y = iris['Species']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

print('The shape of x_train is: {}'.format(x_train.shape))
print('The shape of x_test is: {}'.format(x_test.shape))
print('The shape of y_train is: {}'.format(y_train.shape))
print('The shape of y_test is: {}'.format(y_test.shape))
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
pred_train = model.predict(x_train)
cm = confusion_matrix(y_train, pred_train)

print('confusion matrix decision tree\n', cm)
print('')

print('Akurasi pada saat training: {}'.format(accuracy_score(y_train, pred_train)))
```

```python
pred_test = model.predict(x_test)
cm = confusion_matrix(y_test, pred_test)
accuracy = accuracy_score(y_test, pred_test)
precision = precision_score(y_test, pred_test, average='micro')
recall = recall_score(y_test, pred_test, average='micro')
f1 = f1_score(y_test, pred_test, average='micro')

print('confusion matrix decision tree\n', cm)
print('')

print('Akurasi pada saat testing: %.3f' %accuracy)
print('Presisi pada saat testing: %.3f' %precision)
print('Recall pada saat testing: %.3f' %recall)
print('F1-score pada saat testing: %.3f' %f1)
```
