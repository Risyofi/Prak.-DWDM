```python
import numpy as np
import pandas as pd
from apyori import apriori

df = pd.read_csv("store_data.csv")
df.head(7)
```

```python
df.shape
```

```python
records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i, j]) for j in range(0, df.shape[1])])
print(np.vstack(records))
```

```python
asc_rule = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
for item in asc_rule:
    pair = item[0]
    items = [x for x in pair]

    print("Rule: " + items[1] + " -> " + items[0])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("===========================================")
```
